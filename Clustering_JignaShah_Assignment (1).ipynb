# Clustering Assignment

### Q1 What is unsupervised learning in the context of machine learning?
Unsupervised learning is a machine‑learning technique where the model learns patterns from **unlabeled data**, discovering structure without predefined output.


### Q2 How does K-Means clustering algorithm work?
K-Means works by: (1) randomly initializing k centroids, (2) assigning points to the nearest centroid, (3) recomputing centroids, (4) repeating until convergence.


### Q3 Explain the concept of a dendrogram in hierarchical clustering?
A dendrogram is a **tree-like diagram** used in hierarchical clustering to visualize how clusters merge at each step.


### Q4 What is the main difference between K-Means and Hierarchical Clustering?
K-Means partitions data into k clusters using centroids; hierarchical clustering builds clusters step‑by‑step (bottom‑up or top‑down).


### Q5 What are the advantages of DBSCAN over K-Means?
DBSCAN handles arbitrary shapes, does not require specifying k, and can detect noise; K-Means fails at these.


### Q6 When would you use Silhouette Score in clustering?
Silhouette Score measures how well points fit within their assigned cluster vs. other clusters; used for evaluating clustering quality.


### Q7 What are the limitations of Hierarchical Clustering?
It is computationally expensive, sensitive to noise/outliers, and does not scale well.


### Q8 Why is feature scaling important in clustering algorithms like K-Means?
Feature scaling ensures all features contribute equally; K-Means uses distance which gets distorted without scaling.


### Q9 How does DBSCAN identify noise points?
DBSCAN marks points as noise if they have **fewer than `min_samples` neighbors within `eps` distance**.


### Q10 Define inertia in the context of K-Means?
Inertia is the sum of squared distances of samples to their nearest centroid; lower inertia means tighter clusters.


### Q11 What is the elbow method in K-Means clustering?
Elbow method selects k at the point where inertia stops decreasing sharply (the “elbow”).


### Q12 Describe the concept of "density" in DBSCAN?
Density in DBSCAN is defined by number of points (`min_samples`) within a radius (`eps`).


### Q13 Can hierarchical clustering be used on categorical data?
Yes, if an appropriate distance metric (e.g., Hamming) is used.


### Q14 What does a negative Silhouette Score indicate?
Negative silhouette score indicates a data point is placed in the wrong cluster.


### Q15 Explain the term "linkage criteria" in hierarchical clustering?
Linkage criteria determine how distances between clusters are calculated: single, complete, average, ward.


### Q16 Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
Because it assumes equal cluster sizes and densities; fails when clusters vary in shape or density.


### Q17 What are the core parameters in DBSCAN, and how do they influence clustering?
`eps` (neighborhood radius) and `min_samples` (minimum required neighbors). They determine density threshold.


### Q18 How does K-Means++ improve upon standard K-Means initialization?
K-Means++ spreads out initial centroids so initialization is better and convergence is faster.


### Q19 What is agglomerative clustering?
Agglomerative clustering is a bottom‑up hierarchical technique where small clusters merge to form bigger ones.


### Q20 What makes Silhouette Score a better metric than just inertia for model evaluation?
Silhouette score considers both cohesion & separation; inertia only measures compactness.


### Q21 Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot 

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
X,_=make_blobs(n_samples=500,centers=4,random_state=0)
k=KMeans(n_clusters=4).fit(X)
plt.scatter(X[:,0],X[:,1],c=k.labels_);plt.show()
```
